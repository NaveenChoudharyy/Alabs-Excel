```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_theme()
sns.set_palette('tab20c')

import matplotlib as mpl
mpl.style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing dataset


cars = pd.read_csv("C:/Users/navee/OneDrive/Desktop/Casestudy/regression_car_price_pred_data/quikr_car.csv")

cars.head()
```

```python
cars.drop_duplicates(inplace=True)
```

```python
cars.info()
```

```python
for i in cars.columns:
    print(cars[i].unique())
    print('______________________________________________')
```

```python
cars.isna().sum()
```

```python
for i in cars.columns.difference(['name']):
    print(cars[i].value_counts())
    print('__________________________________________________________')
    print('__________________________________________________________')
```

```python
cars.columns = cars.columns.str.capitalize()

cars.columns
```

```python
cars.loc[~cars.Year.str.isnumeric(), 'Year'] = np.NaN
```

```python
cars.Year.unique()
```

```python
cars.Price = cars.Price.str.replace(',', '')
```

```python
cars = cars[~(cars.Price == 'Ask For Price')]

cars
```

```python
cars = cars[~(cars.Year.isna() & cars.Kms_driven.isna() & cars.Fuel_type.isna())]

cars
```

```python
cars.Kms_driven = cars.Kms_driven.str.replace(',', '')
cars.Kms_driven = cars.Kms_driven.str.replace(' kms', '')
cars.Kms_driven
```

```python
cars.loc[~cars.Kms_driven.str.isnumeric(), 'Fuel_type'] = 'Petrol'


cars.loc[~cars.Kms_driven.str.isnumeric(), 'Fuel_type']
```

```python
cars.loc[~cars.Kms_driven.str.isnumeric(), 'Kms_driven'] = np.NAN
```

```python
cars.Year = cars.Year.astype('int')
```

```python
cars.Price = cars.Price.astype('int')
```

```python
cars.Fuel_type.fillna(cars.Fuel_type.mode()[0], inplace=True)
```

```python
cars.Kms_driven.fillna(int(cars.Kms_driven.astype(float).mean()), inplace=True)
```

```python
cars.Kms_driven = cars.Kms_driven.astype(int)
```

```python
cars.drop(columns='Name', inplace=True)
```

```python
cars.info()
```

```python
cars.select_dtypes([int, float]).describe()
```

```python
cars.select_dtypes([object, 'string']).describe()
```

```python
cars.Price.skew()
```

```python
np.log(cars.Price).skew()
```

```python
cars.Price = np.log(cars.Price)
```

```python
cars.Kms_driven = cars.Kms_driven.clip(upper=cars.Kms_driven.quantile(0.95), lower=cars.Kms_driven.quantile(0.05))

cars.Kms_driven
```

```python
dummies = pd.get_dummies(cars[['Company', 'Year', 'Fuel_type']], drop_first=True).astype(int)
```

```python
cars = pd.concat([dummies,cars[['Kms_driven', 'Price']]], axis=1)
```

```python
cars
```

```python
x_var = list(cars.columns.difference(['Price']))
y_var = 'Price'
```

```python
cars[x_var]
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
skb = SelectKBest(score_func=f_regression, k=20)

skb.fit(cars[x_var], cars[y_var])
```

```python
skb_list = skb.get_feature_names_out()
```

```python
form_0 = 'Price ~ ' + ' + '.join(skb_list)
```

```python
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
Y, X = dmatrices(form_0, cars, return_type='dataframe')
```

```python
vif['Features'] = X.columns
vif['VIF_factor'] = [variance_inflation_factor(X.values, i ) for i in range(X.shape[1])]
```

```python
vif
```

```python
list_1 = list(vif.iloc[1:,: ].loc[vif.VIF_factor <= 4 , 'Features'])

print(list_1)
print(len(list_1))
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(cars[list_1], cars[y_var], test_size=0.3, random_state=42)
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
std_train = std.fit_transform(train_x)
```

```python
std_test = std.transform(test_x)
```

```python
train_x = pd.DataFrame(std_train, columns=train_x.columns)
test_x = pd.DataFrame(std_test, columns=test_x.columns)
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'fit_intercept':[True, False], 'copy_X':[True, False],'positive':[True, False]}
```

```python
lr = GridSearchCV(LinearRegression(), param_grid, cv=5, n_jobs=-1, verbose=True)

lr
```

```python
lr.fit(train_x, train_y)
```

```python
lr.best_params_
```

```python
lr_pred_train = pd.DataFrame({'y_actual': np.exp(train_y), 'y_pred': np.exp(lr.predict(train_x))})

lr_pred_train
```

```python
lr_pred_test = pd.DataFrame({'y_actual': np.exp(test_y), 'y_pred': np.exp(lr.predict(test_x))})

lr_pred_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error, r2_score
```

```python
r2_score(lr_pred_test.y_actual, lr_pred_test.y_pred)
```

```python
print('For Train data')
print('r2_score', r2_score(lr_pred_train.y_actual, lr_pred_train.y_pred))
print('mean_absolute_percentage_error', mean_absolute_percentage_error(lr_pred_train.y_actual, lr_pred_train.y_pred))
print('mean_absolute_error' ,mean_absolute_error(lr_pred_train.y_actual, lr_pred_train.y_pred))
print('mean_squared_error', mean_squared_error(lr_pred_train.y_actual, lr_pred_train.y_pred))
print('root_mean_squared_error', np.sqrt(mean_squared_error(lr_pred_train.y_actual, lr_pred_train.y_pred)))
```

```python
print('For Test data')
print('r2_score', r2_score(lr_pred_test.y_actual, lr_pred_test.y_pred))
print('mean_absolute_percentage_error', mean_absolute_percentage_error(lr_pred_test.y_actual, lr_pred_test.y_pred))
print('mean_absolute_error' ,mean_absolute_error(lr_pred_test.y_actual, lr_pred_test.y_pred))
print('mean_squared_error', mean_squared_error(lr_pred_test.y_actual, lr_pred_test.y_pred))
print('root_mean_squared_error', np.sqrt(mean_squared_error(lr_pred_test.y_actual, lr_pred_test.y_pred)))
```

# KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
param_grid_knn  = {'n_neighbors': [3,4,5,7,10], 'weights':['uniform', 'distance']}
```

```python
knn = GridSearchCV(KNeighborsRegressor(),param_grid=param_grid_knn, cv=5, n_jobs=-1, verbose=True)
```

```python
knn.fit(train_x, train_y)
```

```python
knn_pred_train = pd.DataFrame({'y_actual': np.exp(train_y), 'y_pred': np.exp(knn.predict(train_x))})

knn_pred_train
```
