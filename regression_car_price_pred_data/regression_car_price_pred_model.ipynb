## Variable selection

#### Significant variabls

```python
# Using selectKBest

from sklearn.feature_selection import SelectKBest, f_classif
```

```python
# Getting top 10 significant variables using SelectKBest

skb = SelectKBest(score_func=f_classif, k=40).fit(df[x_var], df[y_var])

skb
```

```python
# Extracting important features name


list_skb = list(skb.get_feature_names_out())

print(list_skb)
```

```python
list_sig_vars = list_skb.copy()
```

#### Removing multi-collinearity

```python
# Using correalion
```

```python
df[list_sig_vars].corr()
```

```python
# Visualization

plt.figure(figsize=(6,6))

sns.clustermap(df[list_sig_vars].corr())

plt.tight_layout()
plt.show()



# It is very difficult to understand the collinerity with corr() as most of the columns doesn't seem co-related here.
```

```python
# Using VIF

from statsmodels.stats.outliers_influence import variance_inflation_factor
from patsy import dmatrices
```

```python
# Creating formulate to be use in dmatrices


form_0 = y_var + ' ~ ' + ' + '.join(list_sig_vars)


print(form_0)
```

```python
# Using dmatrices


Y, X = dmatrices(form_0, df, return_type='dataframe')
```

```python
# Creating dataframe to be stored in the Dataframe

vif = pd.DataFrame()
```

```python
vif['Features']  = X.columns
vif['Vif_factor'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
```

```python
vif
```

```python
# In this all the features seems to be having very small corelation

# Getting the features out as a list


list_sig_vars = list(vif.iloc[1:, :].loc[vif.Vif_factor <=4, 'Features'])

print(len(list_sig_vars), ',', list_sig_vars)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[list_sig_vars], df[y_var], random_state=42, train_size=0.7)
```

# Machine learning models

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { 'fit_intercept': [True, False], 'positive': [True, False], 
              'copy_X': [True, False] }
```

```python
lr = GridSearchCV(LinearRegression(), param_grid, cv=5, verbose=True, n_jobs=-1)
```

```python
lr.fit(x_train, y_train)
```

```python
lr_pred = pd.DataFrame({'y_actual': np.exp(y_train),'y_pred':np.exp(lr.predict(x_train))})

lr_pred
```

### Scoring

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
print('root_mean_squared_error', np.sqrt(mean_squared_error(lr_pred.y_actual, lr_pred.y_pred)))
print('mean_squared_error', mean_squared_error(lr_pred.y_actual, lr_pred.y_pred))
print('mean_absolute_error', mean_absolute_error(lr_pred.y_actual, lr_pred.y_pred))
print('mean_absolute_percentage_error', mean_absolute_percentage_error(lr_pred.y_actual, lr_pred.y_pred))
```

```python
r2_score(lr_pred.y_actual, lr_pred.y_pred)
```
